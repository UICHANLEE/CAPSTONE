```python
# 조건 사전 정의
condition_dict = {
    '년도 및 분기': df['년도 및 분기'],
    '회사명': df['회사명'],
    '학력': df['학력'],
    '과': df['과'],
    '지원 직종': df['지원 직종']
}
```

```python
import itertools

# 조건의 모든 조합 생성
all_combinations = []
for r in range(1, len(condition_dict) + 1):
    combinations_object = itertools.combinations(condition_dict.keys(), r)
    combinations_list = list(combinations_object)
    all_combinations.extend(combinations_list)
```

```python
def magic_lamp(conditions=[], specs=[], metric="mean"):
    if not conditions and not specs:
        agg_dict = {col: metric for col in spec_list}
        return df.agg(agg_dict).to_dict()
    else:
        if conditions:
            query_conditions = " & ".join([f"`{조건_dict[cond]}` == '{cond}'" for cond in conditions])
            condition_df = df.query(query_conditions)
        else:
            condition_df = df
        
        if specs:
            agg_dict = {col: metric for col in specs}
        else:
            agg_dict = {col: metric for col in spec_list}
        
        if condition_df.empty:
            return {}
        
        result = condition_df.agg(agg_dict).to_dict()
        result['value_count'] = condition_df.shape[0]

        return result
    
results = []

# 각 조건에 대해 magic_lamp 함수 호출 및 결과 저장
for condition in final_list:
    result = magic_lamp(condition, [])
    if result:  # result가 비어있지 않을 때만 추가
        result_row = {col: round(result.get(col, None), 2) for col in spec_list}
        result_row['condition'] = " & ".join(condition)
        result_row['value_count'] = result.get('value_count', 0)
        result_row = convert_levels(result_row)
        results.append(result_row)

# 결과를 데이터프레임으로 변환
results_df = pd.DataFrame(results, columns=['condition', 'value_count'] + spec_list)

# 결과를 CSV 파일로 저장
results_df.to_csv('magic_lamp_results.csv', index=False, encoding='cp949')

print("결과가 'magic_lamp_results.csv' 파일로 저장되었습니다.")
```
